```python
import pandas as pd

# Load your dataset
data = pd.read_csv("/Users/pernebayarailym/Documents/Portfolio Projects AP/Python Projects/Project_21_DataScience_Netflix/file.csv")

# Strip leading and trailing whitespace from the date strings
data['Release_Date'] = data['Release_Date'].str.strip()

# Convert to datetime using the specified format, coercing errors to NaT
data['Date_N'] = pd.to_datetime(data['Release_Date'], format='%B %d, %Y', errors='coerce')

# Print the first few rows of the dataframe to check the result
print(data.head())
```
